```python
import numpy as np
import pandas as pd
```

```python
#dataframes : calculations by column

#sum of columns and set to other columns

n_array = np.arange(50).reshape(10,5)
df = pd.DataFrame(n_array)

df.loc[:,'PCT_UNDER_VOLTAGE_MINUTES'] = df.iloc[:,lambda df: [1,2,4]].sum(axis =1)

df
```

```python
#reordering of columns in pandas

df.reindex(columns=[4,0,'PCT_UNDER_VOLTAGE_MINUTES',3,2 ,1 ])
```
