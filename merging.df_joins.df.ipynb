```python
#JOINS ON DATAFRAMES
```

```python
cars = pd.read_csv('datasets/cars_mk.csv')
orders = pd.read_csv('datasets/order.csv')
```

```python
joined_df = pd.concat([cars,orders] , ignore_index=True)   #joins df one over other
joined_df

column_joined_df =pd.concat([cars,orders] ,axis=1)   #joins df along the columns , but does not match the joins like sql
                                                    #see carId columns in 'column_joined_df'
column_joined_df         
```

```python
sql_likejoin = pd.merge(cars,orders)
sql_likejoin

#when there are multiple columns that are common and joins can be performed on them ,we can use 'on'
sql_on_merge = pd.merge(cars, orders , on='CarID')
sql_on_merge      #some of the records missed because they don't have a corespondn=inng carid in orders df
                  #we can make those records appear with using 'how' attribute
    
sql_how_l = pd.merge(cars, orders , on='CarID' , how='left')
sql_how_r = pd.merge(cars, orders , on='CarID' , how='right')
sql_how_r
```
